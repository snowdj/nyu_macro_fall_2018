```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import bisect
```

First let's define a "Market" as a class.  It contains parameters and a supply and demand curve

```python
class Market:
    
    def __init__(self, α=0.1, β=1, γ=1, δ=1):
        self.α, self.β, self.γ, self.δ = α, β, γ, δ

    def qs(self, p):
        return np.exp(self.α * p) - self.β
    
    def qd(self, p):
        return self.γ * p**(-self.δ)
    
```

Here's a function to compute the equilibrium:

```python
def compute_equilibrium(mkt):
    # Define excess demand function
    def h(p):
        return mkt.qd(p) - mkt.qs(p)
    # Find zero of h
    p_star = bisect(h, 2, 4)
    q_star = mkt.qs(p_star)
    
    print(f'Equilibrium price is {p_star: .2f}')
    print(f'Equilibrium quantity is {q_star: .2f}')
```

```python
mkt = Market()
```

```python
compute_equilibrium(mkt)
```

Here's a function for plotting.

```python
def plot_equilibrium(mkt):
    grid = np.linspace(2, 4, 100)
    fig, ax = plt.subplots(figsize=(8, 6))

    ax.plot(grid, mkt.qd(grid), 'b-', lw=2, label='demand')
    ax.plot(grid, mkt.qs(grid), 'g-', lw=2, label='supply')

    ax.set_xlabel('price', fontsize=14)
    ax.set_ylabel('quantity', fontsize=14)
    ax.legend(loc='upper center')

    plt.show()
```

```python
plot_equilibrium(mkt)
```

```python
mkt.α = 0.12
```

```python
compute_equilibrium(mkt)
```

```python
plot_equilibrium(mkt)
```
